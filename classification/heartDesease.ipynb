# Heart Disease

```python
# pip install ucimlrepo
from ucimlrepo import fetch_ucirepo 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features 
y = heart_disease.data.targets 
```

## We start cleaning the data

```python
X.info()
```

## We have few variables so we can give ourselves the freedom to review them one by one with graphs.

## Age


### -Age Distribution 

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X['age'], y, edgecolors='k', s=50)
plt.colorbar(scatter, label='Target')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Target')
plt.show()
```

### -Value Counts of age

```python
age_counts = X['age'].value_counts()
plt.figure(figsize=(12, 6))
age_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Age Value Counts')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
